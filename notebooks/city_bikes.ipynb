# CityBikes

# Send a request to CityBikes for the city of your choice. 

```python
# Importing required libraries
import os   #used to access the values of environment variables
import requests #send api requests

# No API key therefore this is not needed

# Send an API request to Citybikes. Return is a json file. 
url = "https://api.citybik.es/v2/networks/mobibikes"
#url = "http://api.citybik.es/v2/networks"
response = requests.get(url)

#print(response.text)
```

# Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
# Import required librairies and modules
import json # to read json files
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# assign API return to a variable
json_file = json.loads(response.text)

# Finding the normalize statement that will be used to put the json in a dataframe
pd.json_normalize(json_file['network'], record_path=['stations'])
```

# Put your parsed results into a DataFrame.

```python
# Putting json to a dataframe
citybikes_df = pd.json_normalize(json_file['network'], record_path=['stations'])

#Keeping only the relevant columns. Keeping id to count the number of stations per city
citybikes_df.drop(['timestamp', 'extra.bank_card', 'extra.status', 'extra.uid'], axis = 1, inplace=True)

citybikes_df.head()
```

```python
# Exporting our results to a CSV so we can use that data in other parts of the project.
citybikes_df.to_csv('Vancouver_BikeStation.csv', index=True) 
```
